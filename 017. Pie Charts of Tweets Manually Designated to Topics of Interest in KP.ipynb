```python
#Create dataframe of manually assigned topics by number of tweets
d = {'topics': ["service delivery", "keywords assigned", "topic outside keywords", "no topic (i.e. new profile pic)"], 'count': [353521, 2157901, 599498, 3038861]}
df = pd.DataFrame(data=d)
```

```python
#Create a pie chart of assigned topics

df["percent"] = (df["count"]/6149781)*100

import matplotlib.pyplot as plt

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df["topics"]
sizes = df["count"]
explode = (0.1, 0, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
ax1.set_title("Categories Identified in Former Tribal Area Tweets", pad = 25)

plt.savefig("Piechart of Major Categories in Former Tribal Area Tweets.pdf", dpi=300, bbox_inches ="tight")
```
